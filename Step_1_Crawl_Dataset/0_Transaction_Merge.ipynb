```python
df_y = yf.Ticker("BTC-USD").history(
        start= pd.to_datetime(df.Timestamp.head(1).values[0], format='%Y-%m-%d'), end=  pd.to_datetime(df.Timestamp.tail(1).values[0], format='%Y-%m-%d'),
        interval = "1d",
        # group_by = 'ticker',
        # auto_adjust = True,
        # prepost = True,
        # threads = True,
        proxy = None)
```

```python
df_y = df_y[["Close"]]
df_y = df_y.reset_index(level=0)
df_y.head(5)
```

```python
df_y.info()
```

```python
df_y.Date = pd.to_datetime(df_y.Date).dt.tz_localize(None)
```

```python
df.rename(columns = {'Timestamp':'Date', 'miners-revenue':'X1', 'difficulty':'X2', 'hash-rate':'X3', 
                     'median-confirmation-time':'X4', 'avg-block-size':'X5', 'n-unique-addresses':'X6', 
                     'n-transactions-per-block':'X7', 'n-transactions':'X8', 'cost-per-transaction-percent':'X9',
                     'Close':'Y'}, inplace = True)
df = df.merge(df_y, how="inner", on='Date')
df.rename(columns = {'Close':'Y'}, inplace = True)
```
