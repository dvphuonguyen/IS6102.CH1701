<h1>Create a function to scrape data

```python
#"https://coinmarketcap.com/historical/20140914/"
def scrape(date ="20130429"):
  URL = "https://coinmarketcap.com/historical/" + date + "/"
  webpage = requests.get(URL, verify=False)
  soup = BeautifulSoup(webpage.text, "html.parser")
  tr = soup.find_all("tr", attrs = {"class": "cmc-table-row"})
  # print(tr)
  for row in tr:
    name_column = row.find("td", attrs={"class": "cmc-table__cell cmc-table__cell--sortable cmc-table__cell--left cmc-table__cell--sort-by__symbol"})
    # print(name_column)
    if name_column is None:
          break  
    # crypto_name = name_column.find("td", attrs = {"class" : "cmc-table__cell cmc-table__cell--sortable cmc-table__cell--left cmc-table__cell--sort-by__symbol"})#.text.strip()
    crypto_name = name_column.find('div', class_='').text.strip()
    # print(crypto_name)
    cryto_name_list.append(crypto_name)


```

```python
def crawl_data_close(type = "BTC-USD", date = "2014-09-17"):
    current_date = datetime.strptime(date, '%Y-%m-%d')
    next_date = current_date + timedelta(days = 1)
    df = yf.Ticker(type).history(
        start=current_date.strftime("%Y-%m-%d"), end=next_date.strftime("%Y-%m-%d"),
        interval = "1d",
        # group_by = 'ticker',
        # auto_adjust = True,
        # prepost = True,
        # threads = True,
        proxy = None)
    # return df.Close.values[0]
    return df.Close.values[0] if df.Close.values.size != 0 else "Undefined"
    
```

```python
def crawl_data_volume(type = "BTC-USD", date = "2014-09-17"):
    current_date = datetime.strptime(date, '%Y-%m-%d')
    next_date = current_date + timedelta(days = 1)
    df = yf.Ticker(type).history(
        start=current_date.strftime("%Y-%m-%d"), end=next_date.strftime("%Y-%m-%d"),
        interval = "1d",
        # group_by = 'ticker',
        # auto_adjust = True,
        # prepost = True,
        # threads = True,
        proxy = None)
    return df.Volume.values[0] if df.Volume.values.size != 0 else "Undefined"
    # return df.Volume.values[0]
    
```
