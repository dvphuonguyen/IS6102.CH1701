```python
gold_temp = 0
oil_temp = 0
for index, row in df.iterrows():
    if row['Gold_Close'] != "Undefined":
      gold_temp = row['Gold_Close']
    else:
      df.at[index,'Gold_Close'] = gold_temp
    if row['Oil_Close'] != "Undefined":
      oil_temp = row['Oil_Close']
    else:
      df.at[index,'Oil_Close'] = oil_temp
```

```python
# df.to_csv("check.csv")
```

```python
df["Sum_Volume"] = df["BTC_Volume"] + df["Top2_Volume"] + df["Top3_Volume"] + df["Top4_Volume"] + df["Top5_Volume"]
df["Volume_Top2"] = df["Top2_Volume"] / df["Sum_Volume"] 
df["Volume_Top3"] = df["Top3_Volume"] / df["Sum_Volume"] 
df["Volume_Top4"] = df["Top4_Volume"] / df["Sum_Volume"]  
df["Volume_Top5"] = df["Top5_Volume"] / df["Sum_Volume"]  

```

```python
df = df.drop([
        "BTC_Volume",
        "Top2_Volume",
        "Top3_Volume",
        "Top4_Volume",
        "Top5_Volume",
        "Sum_Volume"
        ], axis = 1)
```
